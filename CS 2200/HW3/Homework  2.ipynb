```python
# Problem 1
def reverse_list(l):
    if (len(l) == 0):
        return []
    else:
        return [l[-1]] + reverse_list(l[:-1])

# Problem 1 Test Cases:
print(reverse_list([]))
print(reverse_list([1,2,3,4,5,6]))
print(reverse_list([0]))
print(reverse_list([4,2,5,6,7,8,1,3,4,5]))
print(reverse_list([1,2,3,4,5,6,7,8,9,0,1,2,3,4,5]))
```

```python
# Problem 2    
def srev(l):
    if (len(l) == 0):
        return []
    else :
        a = l[-1]
        if type(a) is list:
            return [srev(a)] + srev(l[:-1])
        else:
            return [l[-1]] + srev(l[:-1])

# Problem 2 Test Cases:
print(srev([1, [5,4], 2, [3,4]]))
print(srev([]))
print(srev([1,2,3,4,5,6,7]))
print(srev([1,[[3,4,6,7],4,3,[5,6,7]],6])) # Should work now
```

\begin{align*}
f(n) = f(n-1) + n^2/(2n-1)(2n+1)

\begin{align*}
g(n) = n(n+1)/2(2n+1)
\end{align*}

\begin{align*}
 D = {N}
\end{align*}

\begin{align*}
f(n) = g(n)
\end{align*}

f(0) = 0^2/(2(0)-1)(2(0)+1) = 0
\end{align*}

\end{align*}

\end{align*}
So P(1) is true
\begin{align*}

\end{align*}
\begin{align*}
g(2) = 2(2+1)/2(2(2)+1) = 6/10 = f(2)

So P(2) is true

\end{align*}
\begin{align*}

\end{align*}
\begin{align*}
g(n-1) = n^2-n/4n-2
\end{align*}

Step 4: Conclusion

We can conclude that for every n in N, f(n) = g(n) where f(n) and g(n) are defined in the above steps.
