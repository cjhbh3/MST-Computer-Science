```python
# Problem 1a
def Merge(L1, L2):
        if (len(L1) == 0):
            return L2
        elif (len(L2) == 0):
            return L1
        else:
            if (L1[0] > L2[0]):
                return [L2[0]] + Merge(L1, L2[1:])
            elif (L1[0] < L2[0]):
                return [L1[0]] + Merge(L1[1:], L2)
            else:
                return [L1[0]] + [L2[0]] + Merge(L1[1:], L2[1:])
# Problem 1b
def MergeSort(L):
    if (len(L) <= 1):
        return L
    
    left = []
    right = []
    
    for i in range(len(L)):
        if (i % 2 == 0):
            left.append(L[i])
        else:
            right.append(L[i])
    
    left = MergeSort(left)
    right = MergeSort(right)
    
    return Merge(left, right)

# Problem 1a 
print("Merge Function Test")
print(Merge([1,4,5,5], [2, 4, 5, 5]))
print(Merge([1,3,4], [2, 4, 5, 5]))
print(Merge([], [2, 4, 5, 5]))
print(Merge([1,3,4,5], []))

# Problem 1c
print(MergeSort([1, 3, 9, 4, 2, 10, 3, 5]))
print(MergeSort([1, 3, 9, 4, 2, 10, 3, 5, 11, 4, 2, 0, 1]))
print(MergeSort([3, 1]))
print(MergeSort([1]))
print(MergeSort([]))
```
